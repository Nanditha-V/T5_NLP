```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
def text_completion(prompt, max_length=50):
    t5_prepared_text = "complete: " + prompt
    input_ids = tokenizer.encode(t5_prepared_text, return_tensors="pt", max_length=1024, truncation=True)
    with torch.no_grad():
        outputs = model.generate(input_ids, max_length=max_length, num_beams=2, early_stopping=True)
    completed_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return completed_text

# Example usage
input_prompt = "Once upon a time, in a magical kingdom, there lived a young princess named Lily. She had a special power to communicate with animals, which made her the talk of the kingdom. One day, while exploring the enchanted forest, Lily discovered a mysterious book hidden behind a waterfall. As she opened the book, a bright light engulfed her, and she found herself transported to a new world. It was a world of mythical creatures and lost treasures, and Lily knew that this was the beginning of a grand adventure. With courage in her heart, she took her first step forward, not knowing what wonders and challenges awaited her. "
completed_text = text_completion(input_prompt, max_length=200)
print("Input prompt:", input_prompt)
print("Completed text:", completed_text)
```

```python
def text_paraphrasing(text):
    t5_prepared_text = "paraphrase: " + text
    input_ids = tokenizer.encode(t5_prepared_text, return_tensors="pt", max_length=1024, truncation=True)
    with torch.no_grad():
        outputs = model.generate(input_ids, max_length=50, num_beams=2, early_stopping=True)
    paraphrased_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return paraphrased_text

# Example
input_text = "The sun rises in the east and sets in the west."
paraphrased_text = text_paraphrasing(input_text)
print("Input text:", input_text)
print("Paraphrased text:", paraphrased_text)
```

```python
def question_answering(question, context):
    t5_prepared_text = f"question: {question} context: {context}"
    input_ids = tokenizer.encode(t5_prepared_text, return_tensors="pt", max_length=1024, truncation=True)
    with torch.no_grad():
        outputs = model.generate(input_ids, max_length=50, num_beams=2, early_stopping=True)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer

# Example
context = "The capital of France is Paris. It is known for its art, culture, and landmarks like the Eiffel Tower."
question = "What is the capital of France?"
answer = question_answering(question, context)
print("Question:", question)
print("Answer:", answer)
```

```python
def text_correction(text):
    t5_prepared_text = "correct: " + text
    input_ids = tokenizer.encode(t5_prepared_text, return_tensors="pt", max_length=1024, truncation=True)
    with torch.no_grad():
        outputs = model.generate(input_ids, max_length=50, num_beams=2, early_stopping=True)
    corrected_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return corrected_text

# Example
text_with_errors = "Thiss sentance has speling errors."
corrected_text = text_correction(text_with_errors)
print("Original Text:", text_with_errors)
print("Corrected Text:", corrected_text)
```
